```python
def fecha_default():
    now = datetime.now()
    dia_ant = now + timedelta(days=-1)
    if int(dia_ant.month)<=9:
        if int(dia_ant.day)<=9:
            dia_ant = '0' + str(dia_ant.day) + '/' + '0' + str(dia_ant.month) + '/' + str(dia_ant.year)
        else:
            dia_ant = str(dia_ant.day) + '/' + '0' + str(dia_ant.month) + '/' + str(dia_ant.year)
    else:
        if(int(dia_ant.day)<=9):
            dia_ant = '0' + str(dia_ant.day) + '/' + str(dia_ant.month) + '/' + str(dia_ant.year)
        else:
            dia_ant = str(dia_ant.day) + '/' + str(dia_ant.month) + '/' + str(dia_ant.year)
    return dia_ant

def fecha_normal():
    now = datetime.now()
    dia_actual = now 
    if int(dia_actual.month)<=9:
        if int(dia_actual.day)<=9:
            dia_actual = '0' + str(dia_actual.day) + '/' + '0' + str(dia_actual.month) + '/' + str(dia_actual.year)
        else:
            dia_actual = str(dia_actual.day) + '/' + '0' + str(dia_actual.month) + '/' + str(dia_actual.year)
    else:
        if(int(dia_actual.day)<=9):
            dia_actual = '0' + str(dia_actual.day) + '/' + str(dia_actual.month) + '/' + str(dia_actual.year)
        else:
            dia_actual = str(dia_actual.day) + '/' + str(dia_actual.month) + '/' + str(dia_actual.year)
    return dia_actual
   
dia_ant = fecha_default()
dia_actual=fecha_normal()
usuario=getpass.getuser()
```
